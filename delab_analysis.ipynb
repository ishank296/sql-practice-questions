```python
%%sql
select d.department_name,
       c.category_name,
       p.product_name,
       sum(oi.order_item_quantity) total_sold_quantity,
       sum(round(cast(oi.order_item_subtotal as numeric),2)) total_revenue
from orders o join
     order_items oi 
on o.order_id=oi.order_item_order_id 
join products p on p.product_id=oi.order_item_product_id
join categories c on c.category_id=p.product_category_id
join departments d on d.department_id=c.category_department_id
group by 
         p.product_name,
         c.category_name,
         d.department_name
order by total_sold_quantity desc
limit 10
```

### Month Wise Sales Comparison

```python
%%sql
with CTE as (
select 
        cast(date_part('month',o.order_date) as integer) ordered_month
       ,cast(date_part('year',o.order_date) as integer) ordered_year
       ,sum(oi.order_item_quantity) total_sold_quantity
       ,sum(round(cast(oi.order_item_subtotal as numeric),2)) total_revenue
from orders o
join order_items oi on o.order_id=oi.order_item_order_id
group by date_part('month',o.order_date)
        ,date_part('year',o.order_date)
order by ordered_year,ordered_month
)
select concat(c1.ordered_year,'-'
       ,c1.ordered_month) ordered_period1
       ,concat(c2.ordered_year,'-'
       ,c2.ordered_month) ordered_period2
      ,c1.total_sold_quantity
      ,c2.total_sold_quantity
      ,c1.total_revenue
      ,c2.total_revenue
from CTE c1 join CTE c2 
on c1.ordered_month=c2.ordered_month+6 and
   c1.ordered_year<c2.ordered_year 
```

### Month By Month Sales Growth for Top 5 Products

```python
%%sql
with top_5_products as
(
select product_name,
       product_id,
       sum(order_item_quantity) total_quantity_sold,
       round(cast(sum(order_item_subtotal) as numeric),3) total_product_revenue
from order_items join products on order_item_product_id=product_id
    group by product_name ,product_id
order by total_quantity_sold desc, total_product_revenue desc
limit 5
),
sales_by_month_year_product as 
(
select date_trunc('month',order_date) ordered_month,
       order_item_product_id,
       product_name,
       sum(order_item_quantity) total_quantity_sold,
       round(cast(sum(order_item_subtotal) as numeric),3) total_product_revenue
    from order_items 
    join top_5_products on
    order_item_product_id=product_id
    join orders on order_id=order_item_order_id
    group by date_trunc('month',order_date),product_name,order_item_product_id
)
select date(ordered_month),product_name,total_quantity_sold,total_product_revenue
from sales_by_month_year_product
order by ordered_month,product_name
```
